# NVDA DD - August 15th
The majority of the Ethereum network hashing power comes from either Nvidia or AMD GPUs. Around Christmas last year the GPU market was completely blown out of proportion, this was caused by mining. By analysing the hashing power of the network it's possible to gain a greater understanding of sales. 

First we retrieve the networking hashing data from [here](https://etherscan.io/chart/hashrate). Then break the hash power growth into the fiscal quarters and compare that to the stocks reactions on earnings dates.

```python
# Load in networking hashing power timeseries data.
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web

font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 14}

plt.rc('font', **font)

series = pd.read_csv('NetworkHash.csv', sep=',', parse_dates=[0])
series = series.set_index(series.Date)
series = series.drop(columns='UnixTimeStamp')
series = series.drop(columns='Date')
series.plot(figsize=(20,10))
plt.ylabel('HashRate (Ghs/s)')
plt.title('Historical Ethereum network hashing power')
plt.show()
```

```python
# Fiscal 2018 - Q3 (July 30, 2017 - October 29, 2017)
Q32018Start = '2017-07-30'
Q32018End = '2017-10-29'

Q3_2018 = series.loc[Q32018Start:Q32018End] # Isolate data to earnings
Q3_2018 = Q3_2018.subtract(Q3_2018.loc[Q32018Start]) #Normalise the data
Q3_2018.plot(figsize=(20,10))
plt.ylabel('HashRate (Ghs/s)')
plt.show()

Q3_2018HashGrowth = int(Q3_2018.max())
print('Growth = ' + str(Q3_2018HashGrowth) + 'Ghs/s')
```
